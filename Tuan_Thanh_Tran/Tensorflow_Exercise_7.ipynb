```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Data Preparation
data = np.genfromtxt('data.csv', delimiter=',')
X = data[:,0:1]
y = data[:,1:]

# create model
model = tf.keras.Sequential(
    [tf.keras.layers.Dense(units=1, input_shape=[1])])

# initialize parameters
model.layers[0].set_weights([np.array([[-0.34]])*100, np.array([0.04])*50])

# declare optimization method and loss function
model.compile(optimizer='sgd', 
              loss='mean_squared_error')

# training
history = model.fit(X, y, 4, epochs=1000, validation_split=0.25)

# parameters after one epoch
print('weight-bias: \n', model.layers[0].get_weights())
```

```python
# plotting 
fig, axs = plt.subplots(1, 2, figsize=(18,5))
axs[0].plot(history.history['loss'], color='b')
axs[0].set(xlabel='iteration', ylabel='losses')
axs[0].plot(history.history['val_loss'], color='r')
axs[0].set(xlabel='iteration', ylabel='losses')
axs[1].scatter(X, y)
axs[1].plot(np.linspace(0, 10, 100), model.predict(np.linspace(0, 10, 100)))
axs[1].set(xlabel='x', ylabel='y')
```

In linear regression, if weight values are initialized with large values, the outputs and losses will be large too. Therefore the model must be trained in a long time to minimize the loss and get optimal weight values. In this case, it took 1000 epoches to converge to an accepted loss value.  

```python
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
tf.keras.backend.set_floatx('float64')

batch_size = 40
epochs = 800

# Data Preparation
data = np.genfromtxt('iris_full_2class.csv', delimiter=',', skip_header=1)
X = data[:,0:4]
y = data[:,4]

# shuffle
inds = np.arange(X.shape[0])
np.random.shuffle(inds)
X = X[inds]
y = y[inds]

# create model
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, activation='sigmoid', input_shape=[4])])

# initialize parameters
model.layers[0].set_weights([np.array([[-0.1], [0.2], [0.4], [0.2]])*100, np.array([0.1])*50])

# declare optimization method and loss function
model.compile(optimizer='sgd', 
              loss='binary_crossentropy',
              metrics=[keras.metrics.BinaryAccuracy()])

# training
history = model.fit(X, y, batch_size, epochs, validation_split=0.2)
```

```python
# plotting 
fig, axs = plt.subplots(1, 2, figsize=(18,5))
axs[0].plot(history.history['loss'], color='b')
axs[0].set(xlabel='iteration', ylabel='losses')
axs[0].plot(history.history['val_loss'], color='r')
axs[1].plot(history.history['binary_accuracy'], color='b')
axs[1].set(xlabel='iteration', ylabel='binary accuracy')
axs[1].plot(history.history['val_binary_accuracy'], color='r')
```

```python
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
tf.keras.backend.set_floatx('float64')

batch_size = 10
epochs = 5000

# Data Preparation
data = np.genfromtxt('iris_full.csv', delimiter=',', skip_header=1)
X = data[:,0:4]
y = data[:,4:]

# shuffle
inds = np.arange(X.shape[0])
np.random.shuffle(inds)
X = X[inds]
y = y[inds]

# create model
model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(4,)))
model.add(tf.keras.layers.Dense(units=3))
model.add(keras.layers.Activation(keras.activations.softmax))

# initialize parameters
model.layers[0].set_weights([np.random.rand(4,3)*100, np.random.rand(3)*50])


# declare optimization method and loss function
model.compile(optimizer='sgd', 
              loss='sparse_categorical_crossentropy', 
              metrics=[keras.metrics.SparseCategoricalAccuracy()])

# training
history = model.fit(X, y, batch_size, epochs, validation_split=0.2)
```

**Solution:** 

- Using Adam's optimization algorithm with large learning rate allowed us to train models converging efficiently after 300 epoches for Linear Regression example, 30 epoches for Logistic Regression example and about 200 epoches for Softmax Regression example.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Data Preparation
data = np.genfromtxt('data.csv', delimiter=',')
X = data[:,0:1]
y = data[:,1:]

# create model
model = tf.keras.Sequential(
    [tf.keras.layers.Dense(units=1, input_shape=[1])])

# initialize parameters
model.layers[0].set_weights([np.array([[-0.34]])*100, np.array([0.04])*50])

# declare optimization method and loss function
opt = tf.keras.optimizers.Adam(learning_rate=10)
model.compile(optimizer=opt, 
              loss='mean_squared_error')

# training
history = model.fit(X, y, 4, epochs=300, validation_split=0.25)

# parameters after one epoch
print('weight-bias: \n', model.layers[0].get_weights())
```

```python
# plotting 
fig, axs = plt.subplots(1, 2, figsize=(18,5))
axs[0].plot(history.history['loss'], color='b')
axs[0].set(xlabel='iteration', ylabel='losses')
axs[0].plot(history.history['val_loss'], color='r')
axs[0].set(xlabel='iteration', ylabel='losses')
axs[1].scatter(X, y)
axs[1].plot(np.linspace(0, 10, 100), model.predict(np.linspace(0, 10, 100)))
axs[1].set(xlabel='x', ylabel='y')
```
